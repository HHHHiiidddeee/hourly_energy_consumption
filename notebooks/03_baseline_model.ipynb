```python
import sys
sys.path.append('../')

import pandas as pd
import warnings
from src.mypackage.data_preparation import prepare_statistical_data
from src.mypackage.forecasting import baseline_forecast
from src.mypackage.evaluation import print_metrics
from src.mypackage.visualization import plot_forecast

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/raw/PJME_hourly.csv")
train, test = prepare_statistical_data(df)
y_train, y_test = train["PJME_MW"].values, test["PJME_MW"].values
```

```python
y_pred_day = baseline_forecast(y_train, y_test, len_pred=24)
y_pred_week = baseline_forecast(y_train, y_test, len_pred=24*7)
y_pred_month = baseline_forecast(y_train, y_test, len_pred=24*30)
y_pred_year = baseline_forecast(y_train, y_test, len_pred=24*365)
```

```python
print_metrics(test["PJME_MW"], y_pred_day, "Baseline Forecast (1 Day)")
print_metrics(test["PJME_MW"], y_pred_week, "Baseline Forecast (1 Week)")
print_metrics(test["PJME_MW"], y_pred_month, "Baseline Forecast (1 Month)")
print_metrics(test["PJME_MW"], y_pred_year, "Baseline Forecast (1 Year)")
```

```python
plot_forecast(test["PJME_MW"], y_pred_day, test["Datetime"], "Baseline Forecast (1 Day)")
```
