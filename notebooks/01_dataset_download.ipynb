```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("robikscube/hourly-energy-consumption", output_dir="../data/raw/")

print("Path to dataset files:", path)
```

```python
import os

files = [f for f in os.listdir(path) if f.endswith(".csv") or f.endswith(".paruqet")]
paths = [os.path.join(path, f) for f in files]
print("CSV and Parquet files in dataset directory:\n", files)
```

```python
import shutil

# Fix typo: rename .paruqet files to .parquet
for f in files:
    if f.endswith(".paruqet"):
        old_path = os.path.join(path, f)
        new_path = os.path.join(path, f.replace(".paruqet", ".parquet"))
        shutil.move(old_path, new_path)
        print(f"Renamed: {f} -> {f.replace('.paruqet', '.parquet')}")

# Re-read file list after renaming
files = [f for f in os.listdir(path) if f.endswith(".csv") or f.endswith(".parquet")]
paths = [os.path.join(path, f) for f in files]
print("\nUpdated file list:\n", files)
```

```python
import pandas as pd

dataset_metadata = pd.DataFrame(columns=["file_name", "n_rows", "n_columns"])
for i, p in enumerate(paths):
    if p.endswith(".csv"):
        df = pd.read_csv(p)
    elif p.endswith(".parquet"):
        df = pd.read_parquet(p)
    dataset_metadata.loc[i] = [files[i], df.shape[0], df.shape[1]]
print(dataset_metadata)
```

```python
os.makedirs("../data/metadata", exist_ok=True)
dataset_metadata.to_csv("../data/metadata/dataset_metadata.csv", index=False)
```
